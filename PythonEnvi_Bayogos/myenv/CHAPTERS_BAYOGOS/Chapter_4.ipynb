# **Exercises**

**Exercise 1**

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_rate = 1.5 * rate
        overtime_pay = overtime_hours * overtime_rate
        pay = regular_pay + overtime_pay
    return pay

# Test it girl
hours_input = 45
rate_input = 10
calculated_pay = compute_pay(hours_input, rate_input)
print(f"Enter Hours: {hours_input}")
print(f"Enter Rate: {rate_input}")
print(f"Pay: {calculated_pay}")
```

# **CHALLENGE**

Challenge

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score          Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		2.0
    [ 82,  85)		2.5
    [ 78,  81)		2.75
    [ 75,  77)		2.5
    [ 0,   74)		5.0

Adapted from [PEW] E3.3
<br>

```python
def computegrade(score):
    if 98 <= score <= 100:
        return "1.0"
    elif 94 <= score < 98:
        return "1.25"
    elif 90 <= score < 94:
        return "1.5"
    elif 86 <= score < 90:
        return "2.0"
    elif 82 <= score < 86:
        return "2.5"
    elif 78 <= score < 82:
        return "2.75"
    elif 75 <= score < 78:
        return "2.5"
    elif 0 <= score < 75:
        return "5.0"
    else:
        return "Invalid score"

# Allow the user to enter the score
try:
    score_input = float(input("Enter score: "))
    grade = computegrade(score_input)
    print(f"Score: {score_input}, Grade: {grade}")
except ValueError:
    print("Invalid input. Please enter a numerical score.")
```
