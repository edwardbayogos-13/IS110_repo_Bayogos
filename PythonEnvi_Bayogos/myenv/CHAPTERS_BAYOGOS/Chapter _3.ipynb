**Exercise 2**

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

```python
hours_str = input('Enter Hours: ')
rate_str = input('Enter Rate: ')

try:
    hours = float(hours_str)
    rate = float(rate_str)
except ValueError:
    print('Error: Please enter numeric input for hours and rate.')
    exit()

if hours <= 40:
    pay = hours * rate
else:
    pay = (40 * rate) + ((hours - 40) * rate * 1.5)

print(f'Pay: {pay:.1f}')
```

# **CHALLENGE**

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score          Grade
    [ 98, 100]      1.0
    [ 94,  97)		1.25
    [ 90,  93)		1.5
    [ 86,  89)		2.0
    [ 82,  85)		2.5
    [ 78,  81)		2.75
    [ 75,  77)		2.5
    [ 0,   74)		5.0

Adapted from [PEW] E3.3
<br> [link text](https://)

```python
score_str = input('Enter score between 0 and 100: ')

try:
    score = float(score_str)
except ValueError:
    print('Bad Score: Please enter a numeric score.')
    exit()

if score < 0 or score > 100:
    print('Bad Score: Score must be between 0 and 100.')
elif score >= 98:
    grade = 1.0
elif score >= 94:
    grade = 1.25
elif score >= 90:
    grade = 1.5
elif score >= 86:
    grade = 2.0
elif score >= 82:
    grade = 2.5
elif score >= 78:
    grade = 2.75
elif score >= 75:
    grade = 2.5
else:
    grade = 5.0

if 'grade' in locals(): # Check if grade was assigned
    print(f'Grade: {grade}')
```
